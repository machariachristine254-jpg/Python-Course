```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,30,1000)
y1=np.sin(x)
y2=np.exp(x)
y3=np.log(x[1:])
y4=1/y1

```

```python
plt.plot(x, y1,"b")
plt.title("Sine Wave")
plt.grid()
plt.show()
```

```python
plt.plot(x, y2,"b")
plt.title("Exp Plot")
plt.grid()
plt.show()
```

```python
plt.plot(x[1:], y3,"b")
plt.title("Ln Plot")
plt.grid()
plt.show()
```

```python
plt.plot(x[:300], y4[:300],"b")
plt.title("Cosec Plot")
plt.grid()
plt.show()
```
